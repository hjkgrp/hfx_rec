```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

This notebook shows the distribution of %E_corr[(T)] diagnostics, as reported in the original DFA recommender paper.

```python
vss_diagnostics = pd.read_csv('../data/vss_hs_diagnostics.csv').set_index('Unnamed: 0')

plt.hist(vss_diagnostics['ecorr_t'])
plt.xlabel(r'HS $\%E_{corr}[(T)]$')
plt.ylabel('Count')
```

```python
#fraction of structures with less MR character
np.sum(vss_diagnostics['ecorr_t'] > 95) / len(vss_diagnostics)
```

```python
csd_diagnostics = pd.read_csv('../data/csd_hs_diagnostics.csv').set_index('Unnamed: 0')

plt.hist(csd_diagnostics['ecorr_t'])
plt.xlabel(r'HS $\%E_{corr}[(T)]$')
plt.ylabel('Count')
```

```python
#fraction of structures with less MR character
np.sum(csd_diagnostics['ecorr_t'] > 95) / len(csd_diagnostics)
```

```python
fig, ax = plt.subplots(figsize=(8,4), ncols=2)
bins = np.arange(94, 98, 0.5)

ax[0].hist(vss_diagnostics['ecorr_t'], alpha=0.5, bins=bins, weights=np.ones(len(vss_diagnostics))/len(vss_diagnostics))
ax[0].set_title('VSS-452')
ax[0].set_xlabel(r'HS DLPNO-$\%E_{corr}[(T)]$')
ax[0].set_ylabel('Frequency')
ax[0].set_ylim((0, 0.55))

ax[1].hist(csd_diagnostics['ecorr_t'], alpha=0.5, bins=bins, weights=np.ones(len(csd_diagnostics))/len(csd_diagnostics))
ax[1].set_title('CSD-76')
ax[1].set_xlabel(r'HS DLPNO-$\%E_{corr}[(T)]$')
ax[1].set_ylabel('Frequency')
ax[1].set_ylim((0, 0.55))
```
